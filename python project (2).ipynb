```python
# finding q1,q3,and iqr
q1=df["return_policy_days"].quantile(0.25)
q3=df["return_policy_days"].quantile(0.75)
iqr=q3-q1
print(q1,q3,iqr)

#set bowndaries
lower=q1-1.5*iqr
upper=q3+1.5*iqr
print(lower,upper)

#filtering
outliers=df[(df["return_policy_days"]<lower)|(df["return_policy_days"]>upper)]
outliers
```

```python
# Removing Outliers
df.drop(outliers.index, inplace=True)
```
